HANDLE MISSING VALUES

```python
print("Handling missing values..\n")

numeric_cols=df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    if df[col].isnull().sum() >0:
        median_value=df[col].median()
        df[col]=df[col].fillna(median_value)
        null_count=df[col].isnull().sum()
        print(f"Filled {null_count} missing values in {col} with {median_value}")

#for categorical columns
categorical_col=df.select_dtypes(include='object').columns
for col in categorical_col:
    if df[col].isnull().sum() >0:
        mode_val=df[col].mode()[0]
        null_count=df[col].isnull().sum()
        df[col]=df[col].fillna(mode_val)
        print(f"Filled {null_count} rows of {col} with {mode_val}")
```

DETECT & TREAT OUTLIERS

```python
print("Detecting and treating outliers..\n")
orginal_count=len(df)
def winsorize_series(series,lower_quantile=0.01,upper_quantile=0.99):
    lower_bound=series.quantile(lower_quantile)
    upper_bound=series.quantile(upper_quantile)
    return series.clip(lower_bound,upper_bound)
# Define reasonable bounds for each column (more permissive than standard ranges)
column_bounds = {
        'latitude': (-90, 90),  # Valid latitude range
        'longitude': (-180, 180),  # Valid longitude range
        'temperature_celsius': (-50, 60),  # Extreme but possible temperatures
        'wind_kph': (0, 200),  # Very strong wind but possible in hurricanes
        'wind_degree': (0, 360),  # Valid wind direction degrees
        'pressure_mb': (800, 1100),  # Extreme but possible pressure values
        'precip_mm': (0, 500),  # Very heavy rainfall but possible
        'humidity': (0, 100),  # Valid humidity percentage
        'cloud': (0, 100),  # Valid cloud coverage percentage
        'visibility_km': (0, 50),  # Very poor to excellent visibility
        'uv_index': (0, 20),  # Extreme but possible UV index
        'gust_kph': (0, 250),  # Very strong wind gusts
        'air_quality_Carbon_Monoxide': (0, 10000),  # Wide range for air quality metrics
        'air_quality_Ozone': (0, 500),
        'air_quality_Nitrogen_dioxide': (0, 500),
        'air_quality_Sulphur_dioxide': (0, 500),
        'air_quality_PM2.5': (0, 500),
        'air_quality_PM10': (0, 1000),
        'air_quality_us-epa-index': (0, 6)  # EPA index typically 1-6
    }
print("First pass: Apply reasonable bounds to prevent impossible values\n")
for col, (lower, upper) in column_bounds.items():
        if col in df.columns:
            # Count extreme values before capping
            extreme_low = (df[col] < lower).sum()
            extreme_high = (df[col] > upper).sum()
            
            if extreme_low > 0 or extreme_high > 0:
                print(f"  {col}: Capping {extreme_low} below {lower}, {extreme_high} above {upper}")
                df[col] = df[col].clip(lower, upper)
print("Second pass: Winsorize extreme but possible values (1st and 99th percentiles)\n")
columns_to_winsorize = [
        'temperature_celsius', 'wind_kph', 'pressure_mb', 'precip_mm',
        'humidity', 'cloud', 'visibility_km', 'uv_index', 'gust_kph',
        'air_quality_Carbon_Monoxide', 'air_quality_Ozone',
        'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide',
        'air_quality_PM2.5', 'air_quality_PM10'
    ]
for col in columns_to_winsorize:
        if col in df.columns:
           
            q01 = df[col].quantile(0.01)
            q99 = df[col].quantile(0.99)
            extreme_low = (df[col] < q01).sum()
            extreme_high = (df[col] > q99).sum()
            
            if extreme_low > 0 or extreme_high > 0:
                print(f"  {col}: Winsorizing {extreme_low+extreme_high} extreme values "
                      f"(below {q01:.2f}, above {q99:.2f})")
                df[col] = winsorize_series(df[col])
# For air quality index, ensure it's integer and within valid range
if 'air_quality_us-epa-index' in df.columns:
        invalid_aqi = ((df['air_quality_us-epa-index'] < 1) | 
                       (df['air_quality_us-epa-index'] > 6) |
                       (df['air_quality_us-epa-index'].isnull())).sum()
        
        if invalid_aqi > 0:
            print(f"  AQI: Fixing {invalid_aqi} invalid values")
            # Convert to integer and cap at valid range
            df['air_quality_us-epa-index'] = (df['air_quality_us-epa-index']
                                             .fillna(1)
                                             .clip(1, 6)
                                             .round()
                                             .astype(int))
    
print(f"Rows preserved: {len(df)}/{orginal_count} ({len(df)/orginal_count*100:.2f}%)")
```

OPTIMIZATION OF DATA TYPES

```python
df.info()
```

```python
def optimize_data_types(df):
    print("=== DATA TYPE OPTIMIZATION ===")
    original_memory = df.memory_usage(deep=True).sum() / 1024**2
    print(f"Original memory: {original_memory:.2f} MB")
    
    df_opt = df.copy()
    
   
    # 2. CONVERT DATETIME
    df_opt['last_updated'] = pd.to_datetime(df_opt['last_updated'])
    
    # 3. DOWNCAST FLOAT64 â†’ FLOAT32 (50% SAVINGS)
    float_cols = [
        'temperature_celsius', 'wind_kph',
        'pressure_mb', 'precip_mm', 'visibility_km', 'uv_index', 'gust_kph',
        'air_quality_Carbon_Monoxide', 'air_quality_Ozone',
        'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide',
        'air_quality_PM2.5', 'air_quality_PM10'
    ]
    
    for col in float_cols:
        df_opt[col] = pd.to_numeric(df_opt[col], downcast='float')
    
    # 4. OPTIMIZE INTEGER COLUMNS (BIG SAVINGS)
    # humidity: 0-100 â†’ uint8 (1 byte instead of 8)
    df_opt['humidity'] = df_opt['humidity'].astype('uint8')
    
    # cloud: 0-100 â†’ uint8 (1 byte instead of 8)
    df_opt['cloud'] = df_opt['cloud'].astype('uint8')
    
    # wind_degree: 0-360 â†’ uint16 (2 bytes instead of 8)
    df_opt['wind_degree'] = df_opt['wind_degree'].astype('uint16')
    
    # air_quality_us-epa-index: 1-6 â†’ uint8 (1 byte instead of 8)
    df_opt['air_quality_us-epa-index'] = df_opt['air_quality_us-epa-index'].astype('uint8')
    
    # Calculate savings
    optimized_memory = df_opt.memory_usage(deep=True).sum() / 1024**2
    savings = original_memory - optimized_memory
    savings_pct = (savings / original_memory) * 100
    
    print(f"Optimized memory: {optimized_memory:.2f} MB")
    print(f"Memory saved: {savings:.2f} MB ({savings_pct:.1f}%)")
    
    return df_opt
df=optimize_data_types(df)
```

FEATURE ENGINEERING ON last_updated column

```python
df['hour']=df['last_updated'].dt.hour
df['day_of_week']=df['last_updated'].dt.dayofweek
df['month']=df['last_updated'].dt.month
df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)

def get_part_of_day(hour):
    if 5 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 17:
        return 'afternoon'
    elif 17 <= hour < 21:
        return 'evening'
    else:
        return 'night'

df['part_of_day'] = df['hour'].apply(get_part_of_day)
df.drop('last_updated',axis=1,inplace=True)
```
